# observations:
Generally, most of the crimes happened during the evening and the night. The crimes least frequently occurred in the early morning, especially between 3am and 6am.

 The crime occurrences then increased steadily through the day as they peaked at noon time (around 17,000 reported incidents) and in the evening between 6pm and 7pm (around 21,000 reported incidents). The crime frequency then decreased across night time and sharply declined between 12am and 1am.

As for Larceny/Theft Larceny they tend to be at their lowest during the morning hours and peak between 5 and 6pm.


```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
data = pd.read_csv("main-data.csv")

categories = ['LARCENY/THEFT', 'ASSAULT', 'VANDALISM', 'WARRANTS', 'VEHICLE THEFT', 'SUSPICIOUS OCC', 'BURGLARY', 'DRUG/NARCOTIC', 'ROBBERY', 'FRAUD']
data= data[data.category.isin(categories)]

data.head()
```

# Through the Week for all incidents

```python
# Split up our data into groups based upon 'Time if the day'
time_Grouping_ = data.groupby('time')

# Find out the count of occurance for each cat.
time = time_Grouping_['category'].count()
#print(time)

by_time_chart = time.plot(kind="line", figsize=(25,5))

by_time_chart.set_xlabel("Time of the Day")
by_time_chart.set_ylabel("Number of incidents")
plt.grid()
plt.show()
```
